```python
df = df.dropna(axis=0)
df_test = df_test.dropna(axis=0)

train_x = df.drop(['Hardness, HB', 'Materia'], axis=1)
train_y = df['Hardness, HB']
test_y = df_test['Hardness, HB']
test_x = df_test.drop(['Hardness, HB', 'Materia'], axis=1)

scaler = MinMaxScaler()
x_train_normilized = scaler.fit_transform(train_x)
x_test_normilized = scaler.transform(test_x)
y_train_normilized = scaler.fit_transform(train_y.values.reshape(-1, 1))
y_test_normilized = scaler.transform(test_y.values.reshape(-1, 1))
```

```python
model2 = tf.keras.models.load_model('C:/Users/Эдуард/Desktop/задание кафедры/models/new models/model2')
x = scaler.inverse_transform(model2.predict(x_test_normilized))[:, 0]
y = scaler.inverse_transform(y_test_normilized).reshape(1, -1)
```

```python
fig = plt.figure(figsize=(5, 4))
plt.xlabel('Empire, HB')
plt.ylabel('Predicted, HB')
plt.grid(True)
plt.scatter(x, y[0])
plt.plot(y[0], y[0], color='red')
plt.show()
fig.savefig('Target.png')
```

```python
df = df[(df['Hardness, HB'] < df['Hardness, HB'].quantile(0.9)) & (df['Hardness, HB'] > df['Hardness, HB'].quantile(0.1))]
```

```python
reg = LinearRegression()
reg.fit(df[['Youngs modulus, GPa ', 'Ultimate strength, GPa']], df['Hardness, HB'])
```

```python
pol_reg = Pipeline([('poly', PolynomialFeatures(degree=2)), ('linear', LinearRegression(fit_intercept=False))])
pol_reg.fit(df[['Youngs modulus, GPa ', 'Ultimate strength, GPa']], df['Hardness, HB'])
```

```python
x_young = df['Youngs modulus, GPa '].values
y_ultimate = df['Ultimate strength, GPa'].values
z_hardness = df['Hardness, HB'].values

plt.figure(figsize=(10, 8))
axes = plt.axes(projection='3d')
axes.scatter3D(x_young, y_ultimate, z_hardness)
axes.plot_trisurf(x_young, y_ultimate, reg.predict(df[['Youngs modulus, GPa ', 'Ultimate strength, GPa']]), color='red', alpha=0.5)

axes.set_xlabel('Youngs modulus, GPa')
axes.set_ylabel('Ultimate strength, GPa')
axes.set_zlabel('Hardness, HB')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
axes = plt.axes(projection='3d')
axes.scatter3D(x_young, y_ultimate, z_hardness)
axes.plot_trisurf(x_young, y_ultimate, pol_reg.predict(df[['Youngs modulus, GPa ', 'Ultimate strength, GPa']]), color='red', alpha=0.5)

axes.set_xlabel('Youngs modulus, GPa')
axes.set_ylabel('Ultimate strength, GPa')
axes.set_zlabel('Hardness, HB')
plt.show()
```
